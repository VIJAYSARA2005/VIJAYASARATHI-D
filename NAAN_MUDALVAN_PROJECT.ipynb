```python
POPULATION_GROWTH_RATE = 0.02  # 2% annual growth rate
WATER_DEMAND_GROWTH_RATE = 0.05  # 5% annual growth rate
WATER_SUPPLY = 1000  # billion cubic meters
WATER_DEMAND = 500  # billion cubic meters
```

```python
years = np.arange(2023, 2051)
population = np.zeros(len(years))
water_demand = np.zeros(len(years))
water_supply = np.zeros(len(years))
```

```python
population[0] = 1380  # million
for i in range(1, len(years)):
    population[i] = population[i-1] * (1 + POPULATION_GROWTH_RATE)
```

```python
water_demand[0] = WATER_DEMAND  # billion cubic meters
for i in range(1, len(years)):
    water_demand[i] = water_demand[i-1] * (1 + WATER_DEMAND_GROWTH_RATE)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, population/1000, label='Population (billion)')
plt.plot(years, water_demand, label='Water Demand (billion cubic meters)')
plt.plot(years, water_supply, label='Water Supply (billion cubic meters)')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Water Resources Management in India')
plt.legend()
plt.show()
```
